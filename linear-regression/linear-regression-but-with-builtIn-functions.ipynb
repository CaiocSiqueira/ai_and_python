```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("weight-height.csv")
data
```

```python
# Split the data based on gender

data_male = data[data["Gender"] == "Male"]
data_female = data[data["Gender"] == "Female"]
```

```python
# Let's start working with the female data because I'm a gentleman

x = data_female["Height"]
y = data_female["Weight"]
```

```python
# Dispersion of the data 

plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x, y);
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train
```

```python
x_train = np.array(x_train).reshape(-1, 1)
x_train
```

```python
x_test = np.array(x_test).reshape(-1, 1)
```

```python
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
c = regression.intercept_
m = regression.coef_
```

```python
function_y = m*x_train+c
function_y
```

```python
#other way of doing the function_y

function_y2 = regression.predict(x_train)
function_y2
```

```python
plt.scatter(x_train, y_train, color="red")
plt.plot(x_train, function_y, color="black")
plt.xlabel("Height in inches")
plt.ylabel("Weight in pounds")
```

```python
# Now testing with the test data (previous one was with the train data)
function_y3 = regression.predict(x_test)
function_y3.flatten()
```

```python
plt.scatter(x_test, y_test, color="yellow")
plt.plot(x_test, function_y3, color="black")
plt.xlabel("Height in inches")
plt.ylabel("Weight in pounds")
```
