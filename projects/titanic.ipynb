```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
train_data
```

```python
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')
test_data
```

```python
train_data.isna().sum()
```

```python
def gender(x):
    if x == 'female':
        return 1
    else:
        return 0

clean_df = train_data
clean_df['Sex'] = train_data['Sex'].apply(gender)
clean_df = clean_df.dropna(subset=['Embarked'])
clean_df['Age'] = clean_df['Age'].fillna(clean_df['Age'].mean().astype(int))
clean_df.drop(['Cabin', 'Ticket'], axis=1, inplace=True)
clean_df.isna().sum()
```

```python
clean_df['Pclass'].unique()
```

```python
clean_df['Embarked'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder

oneHotEncoder = OneHotEncoder()

feature_array = oneHotEncoder.fit_transform(clean_df[['Embarked', 'Pclass']]).toarray()
feature_array
```

```python
oneHotEncoder.categories_
```

```python
feature_labels = np.array(oneHotEncoder.categories_).ravel()
feature_labels[3] = '1class'
feature_labels[4] = '2class'
feature_labels[5] = '3class'
feature_labels
```

```python
oneHotEncodedEmbarkedClassDf = pd.DataFrame(feature_array, columns=feature_labels)
oneHotEncodedEmbarkedClassDf
```

```python
clean_df.drop(['Embarked', 'Pclass'], axis=1, inplace=True)
clean_df = pd.concat([clean_df, oneHotEncodedEmbarkedClassDf], axis=1)
clean_df
```

```python
clean_df.isna().sum()
```

```python
clean_df = clean_df.dropna()
passengers = clean_df.drop(['Name'], axis=1, inplace=True)
clean_df
```

```python
x_train = clean_df.drop('Survived', axis=1)
y_train = clean_df['Survived']
x_test = clean_df.drop("PassengerId", axis=1).copy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
```

```python
clf = LogisticRegression()
clf.fit(x_train, y_train)
y_pred_log_regression = clf.predict(x_test)
accuracy_log_regression = round(clf.score(x_train, y_train)*100, 2)
print('Accuracy: ' + str(accuracy_log_regression) + "%")
```

```python
clf = SVC()
clf.fit(x_train, y_train)
y_pred_svc = clf.predict(x_test)
accuracy_svc = round(clf.score(x_train, y_train)*100,2)
print('Accuracy:' + str(accuracy_svc) + '%')
```

```python
clf = LinearSVC()
clf.fit(x_train, y_train)
y_pred_linear_svc = clf.predict(x_test)
accuracy_linear_svc = round(clf.score(x_train, y_train) * 100, 2)
print('Accuracy:' + str(accuracy_linear_svc) + '%')
```

```python
clf = KNeighborsClassifier(n_neighbors = 3)
clf.fit(x_train, y_train)
y_pred_knn = clf.predict(x_test)
accuracy_knn = round(clf.score(x_train, y_train) * 100, 2)
print ('Accuracy:' + str(accuracy_knn) + '%')
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
y_pred_decision_tree = clf.predict(x_test)
accuracy_decision_tree = round(clf.score(x_train, y_train) * 100, 2)
print('Accuracy:' + str(accuracy_decision_tree) + '%')
```

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(x_train, y_train)
y_pred_random_forest = clf.predict(x_test)
accuracy_random_forest = round(clf.score(x_train, y_train)*100,2)
print('Accuracy:' + str(accuracy_random_forest) + '%')
```

```python
clf = GaussianNB()
clf.fit(x_train, y_train)
y_pred_gnb = clf.predict(x_test)
accuracy_gnb = round(clf.score(x_train, y_train)*100,2)
print ('Accuracy:' + str(accuracy_gnb) + '%')
```

```python
clf = Perceptron(max_iter=5, tol=None)
clf.fit(x_train, y_train)
y_pred_perceptron = clf.predict(x_test)
accuracy_perceptron = round(clf.score(x_train, y_train)*100,2)
print ('Accuracy:' + str(accuracy_perceptron) + '%')
```

```python
clf = SGDClassifier(max_iter=5, tol=None)
clf.fit(x_train, y_train)
y_pred_sgd = clf.predict(x_test)
accuracy_sgd = round(clf.score(x_train, y_train) * 100, 2)
print ("Train Accuracy: " + str(accuracy_sgd) + '%')
```
